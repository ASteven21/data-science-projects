## Column Definition
1. longitude: A measure of how far west a house is; a higher value is farther west
2. latitude: A measure of how far north a house is; a higher value is farther north
3. housingMedianAge: Median age of a house within a block; a lower number is a newer building
4. totalRooms: Total number of rooms within a block
5. totalBedrooms: Total number of bedrooms within a block
6. population: Total number of people residing within a block
7. households: Total number of households, a group of people residing within a home unit, for a block
8. medianIncome: Median income for households within a block of houses (measured in tens of thousands of US Dollars)
9. medianHouseValue: Median house value for households within a block (measured in US Dollars)
10. oceanProximity: Location of the house w.r.t ocean/sea

## Models to Compare
1. Linear Regression
2. Random Forest
3. SVM for Regression

## Note
#### Linear Regression:

Normality Assumption: The normality assumption is primarily related to the distribution of residuals, not the distribution of input variables. Linear regression models are more concerned with the normality of the residuals for valid statistical inferences.

#### Random Forest:

Normality Assumption: Similar to decision trees, random forests are robust and do not assume normality for input variables or residuals. They are non-parametric and handle a wide range of data distributions.

#### Support Vector Machines (SVM) for Regression:

Normality Assumption: SVMs do not explicitly assume normality for input variables or residuals. They are more concerned with finding a hyperplane that separates different classes or captures the trend in regression.

```python
# One hot encode categorical variables
cal_housing_tf = pd.get_dummies(cal_housing)
cal_housing_tf.head()
```

```python
cal_housing_tf['rooms_bedrooms_ratio'] = (cal_housing_tf.total_rooms / cal_housing_tf.total_bedrooms).round(0)
cal_housing_tf['population_households_ratio'] = (cal_housing_tf.population / cal_housing_tf.households).round(0)
cal_housing_tf['rooms_households_ratio'] = (cal_housing_tf.total_rooms / cal_housing_tf.households).round(0)

cal_housing_tf.head()
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(cal_housing_tf.corr(), annot=True, cmap='crest')
plt.show()
```

```python
# Model Development

X = cal_housing_tf.drop('median_house_value', axis=1)
y = cal_housing_tf['median_house_value']

display(X.head(), y.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

display(f'Training shape: {X_train.shape}',
        f'Test shape: {X_test.shape}')
```

```python
# scale the datasets and construct pipeline to perform the training automatically with the chosen model
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression

# apply kfold to do cross validation
kf = KFold(n_splits=5, shuffle=True, random_state=123)

# instantiate model and transformers
scaler = StandardScaler()
lr = LinearRegression()

# construct pipeline
pipeline_lr = Pipeline([
    ('scaler', scaler),
    ('estimator', lr)
])

param_grid = {
    'estimator__fit_intercept': [True, False]
}

grid_search_lr = GridSearchCV(estimator=pipeline_lr, param_grid=param_grid,
                           cv=kf, scoring='neg_mean_squared_error', 
                           return_train_score=True)

# fit data to model
grid_search_lr.fit(X_train, y_train)
```

```python
# check on the training and validation mse
training_mse_lr = -grid_search_lr.cv_results_['mean_train_score'][grid_search_lr.best_index_]
validation_mse_lr = -grid_search_lr.cv_results_['mean_test_score'][grid_search_lr.best_index_]

display(f'Linear Regression Training MSE: {round(training_mse_lr,0)}',
        f'Linear Regression Validation MSE: {round(validation_mse_lr,0)}')
```

```python
# get best parameters
grid_search_lr.best_params_
```

```python
# get best model
best_model = grid_search_lr.best_estimator_
best_model
```

```python
# predict test set
y_pred_lr = best_model.predict(X_test)
y_pred_lr
```

```python
# this will be used to gather the test results of the models
model_test_results = {}
```

```python
# see mse of both training and testing sets
test_mse_lr = mean_squared_error(y_test, y_pred_lr)

# append result to model_test_results dictionary
model_test_results['linear_regression'] = test_mse_lr

display(f'Linear Regression Training MSE: {round(training_mse_lr,0)}',
        f'Linear Regression Test MSE: {round(test_mse_lr,0)}')
```

```python
# use random forest this time
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state=123)
pipeline_rf = Pipeline([
    ('scaler', scaler),
    ('estimator', rf)
])

param_grid = {
    'estimator__n_estimators': [10, 50, 100, 150],
    'estimator__max_depth': [4, 6, 8, 10]
}

grid_search_rf = GridSearchCV(estimator=pipeline_rf, param_grid=param_grid, 
                              cv=kf, scoring='neg_mean_squared_error',
                              return_train_score=True)

# fit rf model
grid_search_rf.fit(X_train, y_train)
```

```python
best_model_rf = grid_search_rf.best_estimator_
best_model_rf
```

```python
grid_search_rf.best_params_
```

```python
# check on the training and validation mse
training_mse_rf = -grid_search_rf.cv_results_['mean_train_score'][grid_search_rf.best_index_]
validation_mse_rf = -grid_search_rf.cv_results_['mean_test_score'][grid_search_rf.best_index_]

display(f'Random Forest Training MSE: {round(training_mse_rf,0)}',
        f'Random Forest Validation MSE: {round(validation_mse_rf,0)}')
```

```python
train_mses = -grid_search_rf.cv_results_['mean_train_score']
val_mses = -grid_search_rf.cv_results_['mean_test_score']

# Get the hyperparameter values for plotting
param_values = [str(params) for params in grid_search_rf.cv_results_['params']]

# Plot the training and validation mean squared errors
plt.figure(figsize=(10, 6))
plt.plot(param_values, train_mses, label='Training Mean Squared Error', marker='o')
plt.plot(param_values, val_mses, label='Validation Mean Squared Error', marker='o')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
# use params max depth 6, n estimator 100
final_rf = RandomForestRegressor(max_depth=6, n_estimators=100, random_state=123)
final_pipeline_rf = Pipeline([
    ('scaler', scaler),
    ('estimator', final_rf)
])

final_pipeline_rf.fit(X_train, y_train)
```

```python
y_pred_rf_train = final_pipeline_rf.predict(X_train)
```

```python
train_mse_rf = mean_squared_error(y_train, y_pred_rf_train)
train_mse_rf
```

```python
# predict on test set
y_pred_rf_test = final_pipeline_rf.predict(X_test)
```

```python
test_mse_rf = mean_squared_error(y_test, y_pred_rf_test)

# update model_test_results dictionary
model_test_results['random_forest'] = test_mse_rf

display(f'Random Forest Training MSE: {round(train_mse_rf,0)}',
        f'Random Forest Test MSE: {round(test_mse_rf,0)}')
```
